Prime or Not


```python
#to check the number is prime or not
num = int(input("Enter number"))
count=0
for i in range(1,num+1):
  if(num%i==0):
    count+=1
if(count==2):
  print("its a prime number")
else:
  print("not a prime")
```

```python
#program on factorial
digit = 12
fact = 1
for i in range(1,digit+1):
  fact = fact*i
print(fact)
```

```python
#program on palindrome
num = 123
temp = num
reverse = 0
while num>0:
  reminder =num%10
  reverse = reverse*10+reminder
  num = num//10
print(reverse)
if(temp==reverse):
  print("palindrome")
else:
  print("Not a palindrome")
```

```python
#Armstrong number
num = 944
sum = 0
temp = num
l = len(str(num))
#print(l)
for i in range(l+1):
  rem = num%10
  sum = sum+(rem**l)
  num = num//10
print(sum)
if(temp==sum):
  print("Armstrong")
else:
  print("Not a armstrong")
```
