1. Simple Perceptron

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Dataset
X = np.array([[0],[1],[2],[3],[4],[5]])
y = np.array([0,0,0,1,1,1])

# Perceptron
model = Sequential([
    Dense(1, activation='sigmoid', input_shape=(1,))
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X, y, epochs=100, verbose=0)

prediction = model.predict(np.array([[4]]))

print("Prediction:", prediction[0][0])
```

2. Artificial Neural Network (ANN)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

X = np.array([[2],[4],[6],[8],[10]])
y = np.array([0,0,1,1,1])

model = Sequential([
    Dense(8, activation='relu', input_shape=(1,)),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X, y, epochs=100)

prediction = model.predict(np.array([[7]]))

print("Prediction:", prediction[0][0])
```

3. Image Classification using MNIST

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=5)

loss, accuracy = model.evaluate(X_test, y_test)

print("Accuracy:", accuracy)
```

4. Deep Neural Network (DNN)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

X = np.random.rand(100,5)
y = np.random.randint(0,2,100)

model = Sequential([
    Dense(64, activation='relu', input_shape=(5,)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X, y, epochs=20)
```

5. Forward Propagation Example

```python
import numpy as np

inputs = np.array([2,3])

weights = np.array([0.5,0.8])

bias = 1

z = np.dot(inputs, weights) + bias

print("Output:", z)
```

6. ReLU Activation

```python
import tensorflow as tf

values = [-5,-2,0,3,7]

print(tf.nn.relu(values).numpy())
```

7. Sigmoid Activation

```python
import tensorflow as tf

values = [-3,-1,0,2,5]

print(tf.nn.sigmoid(tf.constant(values, dtype=tf.float32)).numpy())
```

8. Softmax Activation

```python
import tensorflow as tf

scores = [3.2,1.5,0.5]

print(tf.nn.softmax(scores).numpy())
```

9. Tanh Activation

```python
import tensorflow as tf

values = [-2,-1,0,1,2]

print(tf.nn.tanh(tf.constant(values, dtype=tf.float32)).numpy())
```

10. Leaky ReLU

```python
import tensorflow as tf

values = [-5,-2,0,2,5]

print(tf.nn.leaky_relu(values).numpy())
```
