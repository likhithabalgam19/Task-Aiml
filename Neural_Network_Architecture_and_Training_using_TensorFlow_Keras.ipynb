1. Forward Propagation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Dataset
X = np.array([[2],[4],[6],[8],[10]])
y = np.array([0,0,1,1,1])

# Model
model = Sequential([
    Dense(8, activation='relu', input_shape=(1,)),
    Dense(1, activation='sigmoid')
])

# Compile
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(X, y, epochs=100, verbose=0)

# Forward Propagation (Prediction)
prediction = model.predict(np.array([[7]]))

print("Probability:", prediction[0][0])

if prediction[0][0] >= 0.5:
    print("Prediction: PASS")
else:
    print("Prediction: FAIL")
```

2. Mean Squared Error (MSE)

```python
from sklearn.metrics import mean_squared_error

actual = [10,20,30]
predicted = [12,18,35]

loss = mean_squared_error(actual, predicted)

print("Mean Squared Error:", loss)
```

3. Binary Cross Entropy

```python
from sklearn.metrics import log_loss

actual = [1,0,1]
prediction = [0.9,0.2,0.8]

loss = log_loss(actual, prediction)

print("Binary Cross Entropy:", loss)
```

4. Categorical Cross Entropy

```python
import tensorflow as tf
from tensorflow.keras.losses import CategoricalCrossentropy
import numpy as np

# One-hot encoded labels
actual = np.array([
    [1,0,0],
    [0,1,0],
    [0,0,1]
])

prediction = np.array([
    [0.8,0.1,0.1],
    [0.2,0.7,0.1],
    [0.1,0.2,0.7]
])

loss_fn = CategoricalCrossentropy()

loss = loss_fn(actual, prediction)

print("Categorical Cross Entropy:", loss.numpy())
```

5. Backward Propagation (Weight Update)

```python
weight = 2.5
gradient = 0.8
learning_rate = 0.1

new_weight = weight - learning_rate * gradient

print("Old Weight:", weight)
print("Gradient:", gradient)
print("New Weight:", new_weight)
```

6. Gradient Descent

```python
weight = 5
gradient = 3
learning_rate = 0.1

for epoch in range(5):
    weight = weight - learning_rate * gradient
    print(f"Epoch {epoch+1}: Weight = {weight:.2f}")
```

7. Complete Neural Network (TensorFlow/Keras)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Dataset
X = np.array([[2],[4],[6],[8],[10]])
y = np.array([0,0,1,1,1])

# Build Model
model = Sequential([
    Dense(8, activation='relu', input_shape=(1,)),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(X, y, epochs=100, verbose=1)

# Evaluate
loss, accuracy = model.evaluate(X, y, verbose=0)

print("Loss:", loss)
print("Accuracy:", accuracy)

# Predict
test = np.array([[3],[5],[7],[9]])

predictions = model.predict(test)

for x, p in zip(test, predictions):
    if p[0] >= 0.5:
        result = "PASS"
    else:
        result = "FAIL"

    print(f"Study Hours = {x[0]}, Probability = {p[0]:.4f}, Prediction = {result}")
```

9. Display Weights and Biases

```python
for i, layer in enumerate(model.layers):

    weights, biases = layer.get_weights()

    print(f"\nLayer {i+1}")

    print("Weights")
    print(weights)

    print("Biases")
    print(biases)
```
