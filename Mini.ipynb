```python
df = pd.read_csv("instagram.csv")
df.head()
```

```python
print(df.isnull().sum())
```

```python
platform_count = df['Platform'].value_counts()

print("Platform count:")
print(platform_count)

plt.figure(figsize=(8,5))

sns.countplot(
    x='Platform',
    data=df,
    order=df['Platform'].value_counts().index,
    palette="Set2"
)

plt.title("Social Media Platform Usage")
plt.xlabel("Platform")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
platform_df = df['Platform'].value_counts().head(10)

print("Top Platforms:")
print(platform_df)

plt.figure(figsize=(8,5))

sns.barplot(
    x=platform_df.values,
    y=platform_df.index,
    palette="Set2"
)

plt.title("Top Social Media Platforms")
plt.xlabel("Count")
plt.ylabel("Platform")

plt.show()
```

```python
stress = df['Stress_Level'].value_counts()

print("Stress Level Count:")
print(stress)

plt.figure(figsize=(8,5))

sns.barplot(
    x=stress.values,
    y=stress.index,
    palette="Set2"
)

plt.xlabel("Number of Users")
plt.ylabel("Stress Level")
plt.title("Stress Level Distribution")

plt.show()
```
