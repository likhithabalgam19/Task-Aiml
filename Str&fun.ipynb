**Strings in Python**

What is a String?


A String is a sequence of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).



Features of Strings

Strings are ordered Strings are immutable (cannot be modified directly) Supports indexing and slicing Can contain letters, numbers, and special characters

**String Indexing**

Each character has a position called an index.

text = "Python" Character P y t h o n
 Positive Index 0 1 2 3 4 5
 Negative Index -6 -5 -4 -3 -2 -1


Function Purpose

upper() Convert to uppercase lower() Convert to lowercase capitalize() First letter uppercase title() Every word first letter uppercase strip() Remove spaces replace('replaceval) Replace text split() Convert string to list find('str') Find position returns index count('char) Count occurrences startswith(value) Check starting text returns bool endswith('value) Check ending text returns bool

```python
s = 'python programming'
print(s)
print(len(s))
print(s.upper())
print(s.lower())
print(s.capitalize())
print(s.title())
print(s.replace('python','java'))
print(s.split())
print(s.find('programming'))
print(s.count('o'))
print(s.startswith("Python"))
print(s.endswith("programming"))
'''s1 = '  python  '
#print(s1)

#print(len(s1))
#print(s1.strip())
#print(s1) '''
```

**Functions in Python**

What is a Function?

A function is a block of reusable code that performs a specific task. Instead of writing the same code multiple times, we can create a function and call it whenever needed.


Benefits of Functions Code Reusability Reduces Code Duplication Improves Readability Easier Debugging and Maintenance

```python
def function_name():
    # Function body
    print("Hello World")

function_name()
```

**Types of Functions**

1.Function Without Parameters and Without Return Value

```python
def display():
    print("Learning Python Functions")

display()
```

2. Function With Parameters

Parameters allow us to pass data to a function.

```python
def greet(name):
    print("Hello", name)

greet("Likhitha")
greet("Abhi")
```

3. Function With Return Value

```python
def add(a, b):
    return a + b

result = add(10, 20)
print(result)
```

4. Function With Parameters and Return Value

```python
def square(num):
    return num * num

result = square(5)
print(result)
```
