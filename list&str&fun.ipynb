**List**

Definition

A List is an ordered, mutable (changeable) collection of items. It allows duplicate values and can store different data types.

Features Ordered Mutable Allows duplicates Supports indexing

```python
fruits = ["Apple", "Banana", "Mango"]

print(fruits)
print(fruits[0])  # Access first element
```

Important Functions Function Description


*   append() Add item at end
*   insert() Insert item at specific position

*   remove() Remove item

*   pop() Remove item using index


*   sort() Sort list


*   reverse() Reverse list



*   len() Count elements
*  count() Count occurrences




```python
 stud = ['nikhitha','kiran','madhu']
 print(stud)
 print(type(stud))
 stud.append("shila")
 print(stud)
 stud.insert(1,'naresh')
 print(stud)
 stud.remove('madhu')
 print(stud)
 stud.pop(0)
 print(stud)
 stud.sort()
 print(stud)
 stud.reverse()
 print(stud)
 l = len(stud)
 print(l)
 fruit = "Banana"
 fruit.count('a')
```


**2. Tuple**





Definition

A Tuple is an ordered and immutable (cannot be changed) collection of items.

Features Ordered Immutable Allows duplicates Faster than lists

```python
colors = ("Red", "Green", "Blue")

print(colors)
print(colors[1])
```

Important Functions Function Description



*   len() Number of elements

*    count() Count occurrence
*   index() Find position





```python
car_brands = ('benz','Thar','suzuki','Audi','Thar')
print(car_brands)
print(len(car_brands))
print(car_brands.count('Thar'))
print(car_brands[2])
```

3. Set

Definition

A Set is an unordered collection of unique elements.

Features Unordered Mutable No duplicates allowed Faster searching

```python
numbers = {1, 2, 3, 4}

print(numbers)
```

Important Functions Function Description



*   add() Add element

*   remove() Remove element

*   union() Combine sets
*  intersection() Common elements


*   difference() Unique elements


*   clear() Remove all elements



```python
name1 = {'srija','shiva','teja','priya','Ajith'}
name2 = {'teja','priya','Ajith','kiran','ashok'}
print(name1)
print(name2)
name1.add('task')
print(name1)
name1.remove('srija')
print(name1)
unions = name1.union(name2)
print(unions)
inter = name1.intersection(name2)
print(inter)
dif = name1.difference(name2)
print(dif)
name1.clear()
print(name1)
```

4. Dictionary

Definition

A Dictionary stores data in key-value pairs.

Features Mutable Keys must be unique Fast lookup Stores related information together

```python
student = {
    "name": "John",
    "age": 21,
    "course": "AIML"
}
print(student)
print(student.keys())
print(student.values())
print(student.items())
print(student.get('name'))
student.pop('name')
print(student)
student.update({'name':'hello'})
print(student)
student.update({'age':23})
print(student)
```

Important Functions Function Description



*   keys() Returns all keys

*   values() Returns all values

*  items() Returns key-value pairs
*   List itemget() Safely access value

*  pop() Remove key


*   update() Update dictionary



Strings in Python

What is a String?

A String is a sequence of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

Features of Strings



*   Strings are ordered

*   Strings are immutable (cannot be modified directly)
*   Supports indexing and slicing


*   Can contain letters, numbers, and special characters



String Indexing


Each character has a position called an index.



*   text = "Python"

*   Character P y t h o n
*   Positive Index 0 1 2 3 4 5


*   Negative Index -6 -5 -4 -3 -2 -1



Function Purpose



*   upper() Convert to uppercase

*  lower() Convert to lowercase

*   capitalize() First letter uppercase
*   title() Every word first letter uppercase
*   strip() Remove spaces
*  replace('replaceval) Replace text


*   split() Convert string to list


*  find('str') Find position returns index


*   count('char) Count occurrences

*  startswith(value) Check starting text returns bool

*   endswith('value) Check ending text returns bool





```python
s = 'python programming'
print(s)
print(len(s))
print(s.upper())
print(s.lower())
print(s.capitalize())
print(s.title())
print(s.replace('python','java'))
print(s.split())
print(s.find('programming'))
print(s.count('o'))
print(s.startswith("Python"))
print(s.endswith("programming"))
'''s1 = '  python  '
#print(s1)

#print(len(s1))
#print(s1.strip())
#print(s1) '''
```

Functions in Python

What is a Function?

A function is a block of reusable code that performs a specific task. Instead of writing the same code multiple times, we can create a function and call it whenever needed.

Benefits of Functions Code Reusability Reduces Code Duplication Improves Readability Easier Debugging and Maintenance

```python
def function_name():
    # Function body
    print("Hello World")

function_name()
```

Types of Functions

1.Function Without Parameters and Without Return Value

```python
def display():
    print("Learning Python Functions")

display()
```

2.Function With Parameters

Parameters allow us to pass data to a function.

```python
def greet(name):
    print("Hello", name)

greet("Likhitha")
greet("Rahul")
```

3.Function With Return Value

```python
def add(a, b):
    return a + b

result = add(10, 20)
print(result)
```

4.Function With Parameters and Return Value

```python


def square(num):
    return num * num

result = square(5)
print(result)

```
