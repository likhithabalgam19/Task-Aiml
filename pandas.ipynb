```python
import pandas as pd

df1 = pd.DataFrame({
    "name": ["Rahul", "Sneha", "Arjun", "Meera"],
    "age": [21, 22, 20, 23]
})

df2 = pd.DataFrame({
    "name": ["Rahul", "Sneha", "Arjun", "Meera"],
    "branch": ["CSE", "ECE", "AIML", "DS"]
})

merged = pd.merge(df1, df2, on="name")

print(merged)
```
